```python
import torch
import seaborn as sns
import matplotlib.pyplot as plt

# Load saved weights
attn_weights = torch.load('results/attn_weights.pt').cpu().detach()
# average across heads
avg_attn = attn_weights[0].mean(dim=0) 

tokens = ["Transformers", "use", "[MASK]", "attention"]
sns.heatmap(avg_attn, xticklabels=tokens, yticklabels=tokens, annot=True, cmap='viridis')
plt.title("Self-Attention Heatmap")
plt.show()
```
